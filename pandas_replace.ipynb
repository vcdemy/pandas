# .replace() => 數值替換

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(987)
data = np.random.randint(1, 11, (5, 5))
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.columns = ['col1', 'col2', 'col3', 'col4', 'col5']
df.index = ['row1', 'row2', 'row3', 'row4', 'row5']
```

```python
df
```

## 一對一取代

```python
df.replace(10, 20)
```

```python
df
```

```python
df.replace(10, 20, inplace=True)
```

```python
df
```

## 一對多取代 => 使用list

```python
df
```

```python
df.replace([1, 2, 3], 15)
```

## 多對多取代 => 使用dict

```python
df
```

```python
df.replace({1:11, 2:12, 3:13})
```
