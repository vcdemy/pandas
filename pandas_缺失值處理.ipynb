# 缺失值處理

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(987)
data = np.random.randint(20, 101, (10, 4))
```

```python
df = pd.DataFrame(data)
df.columns = ['col1', 'col2', 'col3', 'col4']
```

```python
df
```

```python
df.iloc[1, 2] = np.nan
df.iloc[3, 1] = np.nan
df.iloc[3, 3] = np.nan
df.iloc[5, 0] = np.nan
df.iloc[5, 1] = np.nan
df.iloc[7, 0] = np.nan
df.iloc[7, 1] = np.nan
df.iloc[7, 2] = np.nan
df.iloc[7, 3] = np.nan
```

```python
df
```

## .dropna()

```python
df.dropna()
```

```python
df.dropna(subset=['col1'])
```

```python
df.dropna(subset=['col1', 'col3'])
```

```python
df.dropna(how='all')
```

```python
df.dropna(thresh=3)
```

## .fillna()

```python
df
```

```python
df.fillna(0)
```

```python
df.fillna(df.mean())
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='pad')
```

```python
df.fillna(method='bfill')
```
