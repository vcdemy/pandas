# .groupby()

```python
import pandas as pd
import numpy as np
```

## 產生虛擬資料集

```python
keys = np.random.choice(['A','B','C'], 20)
```

```python
keys
```

```python
values = np.random.randint(0, 10, 20)
```

```python
df = pd.DataFrame(zip(keys, values))
```

```python
df
```

```python
df.columns = ['keys','values']
```

```python
df
```

```python
df.sort_values('keys')
```

## .groupby()

```python
grps = df.groupby('keys')
```

```python
grps.get_group('A')
```

```python
grps.agg(['count', sum])
```

```python
tuple(grps)
```

```python
dict(tuple(grps))
```
