# 合併DataFrame

```python
import numpy as np
import pandas as pd
```

```python
df1 =  pd.DataFrame([['a', 1], ['b', 2]], columns=['col1', 'col2'], index=['row1', 'row2'])
```

```python
df2 =  pd.DataFrame([['c', 3], ['d', 4]], columns=['col1', 'col2'], index=['row1', 'row2'])
```

```python
df3 =  pd.DataFrame([['e', 1, 100], ['f', 2, 200]], columns=['col1', 'col2', 'col3'], index=['row1', 'row2'])
```

```python
df4 =  pd.DataFrame([['g', 3, 300], ['h', 4, 400]], columns=['col1', 'col2', 'col3'])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
df4
```

## .concat()

沿著特定軸向合併DataFrame。

https://pandas.pydata.org/docs/reference/api/pandas.concat.html

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df3])
```

```python
pd.concat([df1, df3], join="inner")
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df4], axis=1)
```

## .append()

將一個DataFrame附加到另一個之後。

https://pandas.pydata.org/pandas-docs/version/1.3/reference/api/pandas.DataFrame.append.html

```python
df1.append(df2)
```

```python
df1.append(df2, verify_integrity=True)
```

```python
df1.append(df2, ignore_index=True)
```

```python
df1.append(df3)
```

```python
df1.append(df4)
```

## .join()

會根據index把DataFrame合併在一起。

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, lsuffix='_df1', rsuffix='_df2')
```

```python
df2.columns = ['col3', 'col4']
```

```python
df1.join(df2)
```
